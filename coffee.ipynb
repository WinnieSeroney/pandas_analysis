```python
# Create a new DataFrame called espresso_df that only contains rows where the Product is 'Espresso'.
```

```python
# Sort the original df by Revenue in descending order (highest revenue at the top) and display the result.
```

```python
import pandas as pd
df = pd.read_csv('coffee_shop.csv')
sorted_df = df.sort_values(by='Revenue', ascending=False)
print(sorted_df)
```

```python
# The dataset has Revenue and Units_Sold, but not the price per item. 
# Create a new column called Price by dividing Revenue by Units_Sold.
```

```python
import pandas as pd
df = pd.read_csv('coffee_shop.csv')
df['Price'] = df['Revenue'] / df['Units_Sold']
print(df)
```

```python
# Calculate the total Revenue earned across all days and regions.
```

```python
import pandas as pd
df = pd.read_csv('coffee_shop.csv')
total_revenue = df['Revenue'].sum()
print("Total Revenue:", total_revenue)
```

```python
# Group the data by Region and calculate the total Units_Sold for each region. 
# Which region sold the most?

```

```python
import pandas as pd
df = pd.read_csv('coffee_shop.csv')
region_sales = df.groupby('Region')['Units_Sold'].sum()
top_region = region_sales.idxmax()  
print("Total Units Sold by Region:\n", region_sales)
print("Region with Highest Sales:", top_region)
```

```python
# Filter the DataFrame to show only rows where the Product is 'Latte' AND the Revenue is greater than 130.
```

```python
import pandas as pd
df = pd.read_csv('coffee_shop.csv')
latte_high_revenue = df[(df['Product'] == 'Latte') & (df['Revenue'] > 130)]
print(latte_high_revenue)
```

```python
# Create a simple bar plot showing the total Revenue for each Product.
# (Hint: She can use df.groupby('Product')['Revenue'].sum().plot(kind='bar'))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('coffee_shop.csv')
product_revenue = df.groupby('Product')['Revenue'].sum()
product_revenue.plot(kind='bar', title='Total Revenue by Product')

plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.show()
```
