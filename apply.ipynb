Load the dataset into a DataFrame.

```python
import pandas as pd 
df=pd.read_csv("apply.csv")
print(df)
```

Create a new column called Revenue calculated as:Price × Units_Sold


```python
df['Revenue']=df['Units_Sold']*df['Price']
print(df)
```

Use .apply() with a lambda function to create a new column called Taxed_Revenue by adding 16% VAT to the Revenue.

```python
df['Taxed_Revenue']=df['Revenue'].apply(lambda x: x*1.16 )
print(df)
```

Write a function that:Applies a 10% discount if Units_Sold ≥ 5
 Otherwise keeps the revenue unchanged

```python
def apply_discount(row):
    if row['Units_Sold'] >=5:
        return row['Price']*0.9
    else:
        return row['Revenue'] 
    
df['Discounted_Revenue']=df.apply(apply_discount, axis=1)
print(df)
```

Display only the Product, Revenue, and Discount columns

```python
print(df[['Product', 'Revenue', 'Discounted_Revenue']])
```
