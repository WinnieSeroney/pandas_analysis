```python
import pandas as pd
import numpy as np



data = {
    'Date': ['2023-10-01', '2023-10-01', '2023-10-01', '2023-10-02', '2023-10-02', 
             '2023-10-02', '2023-10-03', '2023-10-03', '2023-10-03', '2023-10-04'],
    'Product': ['Cappuccino', 'Latte', 'Espresso', 'Latte', 'Cappuccino', 
                'Espresso', 'Cappuccino', 'Latte', 'Espresso', 'Latte'],
    'Price': [4.50, 4.00, 3.00, 4.00, 4.50, 3.00, 4.50, 4.00, 3.00, 4.00],
    'Quantity_Sold': [20, 15, 30, 25, 18, 35, 22, 10, 40, 30],
    'Customer_Rating': [4.8, 4.2, 3.9, 4.5, 4.7, 4.0, 4.9, 3.5, 4.1, 4.6]
}


df=pd.DataFrame(data)
print(df)

```

```python
# DataFrame Creation: Convert the data dictionary into a pandas DataFrame called df.
```

```python
import pandas as pd
df=pd.DataFrame(data)
print(df)
```

```python
# Inspection: Display the first 5 rows of the DataFrame and check the data types of each column (using .info()).
```

```python

print(df.head())
print(df.info())    
```

```python
# Basic Statistics: What is the average (mean) Customer_Rating across all sales?
```

```python
average_rating = df['Customer_Rating'].mean()
print("Average Customer Rating:" ,average_rating )
```

```python
# Column Calculation: Create a new column called Revenue. This should be calculated as Price multiplied by Quantity_Sold.
```

```python
df['Revenue'] = df['Price'] * df['Quantity_Sold']
print(df)
```

```python
# Filtering: Create a new DataFrame (or filter the existing one) to show only the transactions where the Customer_Rating was 4.5 or higher.
```

```python
high_rating_df = df[df['Customer_Rating'] >= 4.5]
print(high_rating_df)   
```

```python
# Sorting: Sort the original DataFrame by Quantity_Sold in descending order (highest quantity at the top).
```

```python
df_sorted = df.sort_values(by='Quantity_Sold', ascending=False)
print(df_sorted)
```

```python
# Selection: From the sorted data, display only the Date, Product, and Revenue columns.
```

```python
df_selected= df_sorted[['Date', 'Product', 'Revenue']]
print(df_selected)
```

```python
# Grouping (Aggregation): Calculate the total Revenue generated by each Product. (Hint: Use .groupby()).
```

```python
total_revenue_by_product = df.groupby('Product')['Revenue'].sum().reset_index()
print(total_revenue_by_product) 
```

```python
# Grouping (Aggregation): Calculate the total Revenue generated by each Product. (Hint: Use .groupby()).
```

```python
total_revenue_by_product = df.groupby('Product')['Revenue'].sum()
print(total_revenue_by_product)
```

```python
# Plotting: Create a simple Bar Chart showing the Total Revenue per Product (calculated in question 8).
```

```python
import matplotlib.pyplot as plt
total_revenue_per_product=df.groupby('Product')['Revenue'].sum()
total_revenue_per_product.plot(kind='bar',color='orange')
plt.title('Total Revenue per Product')
plt.xlabel('Product')   
plt.ylabel('Total Revenue')
plt.show()
```
