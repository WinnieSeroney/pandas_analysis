//load the dataset into a pandas data set called pd

```python
import pandas as pd
df=pd.read_csv("sales_dirty.csv")
print(df)
```

```python
Determine the total number of missing values in the entire dataset.
```

```python
df.isnull().sum().sum()
print(df)
```

Identify which column has the highest number of missing values.

```python
df.isnull().sum().idxmax()
```

Remove all rows that contain at least one missing value.

```python
df_no_missing = df.dropna()
print(df_no_missing)    
```

Remove all columns that contain any missing values.

```python
df_no_missing_cols=df.dropna(axis=1)
print(df_no_missing_cols)   
```

Fill missing values in the Price column with the mean price.

```python
mean_price=df['Price'].mean()
df['Price']=df['Price'].fillna(mean_price)  

print(df['Price'])
```

```python
Fill missing values in the Customer_Rating column with the median rating.
```

```python
median_rating = df['Customer_Rating'].median()
df['Customer_Rating'] = df['Customer_Rating'].fillna(median_rating)
print(df['Customer_Rating'])    
```

Fill missing values in the Quantity_Sold column with 0.

```python
df['Quantity_Sold']=df['Quantity_Sold'].fillna(0)
print(df['Quantity_Sold'])
```

Identify duplicate rows in the dataset.

```python
duplicates=df.duplicated()
print(duplicates)   
```

Display all rows that are duplicated.

```python
duplicate_rows=df[df.duplicated()]
print(duplicate_rows)   
```

Remove duplicate rows from the dataset

```python
df_cleaned=df.drop_duplicates()
print(df_cleaned)   
```
