Load the dataset into a DataFrame

```python
import pandas as pd  
df=pd.read_csv("advanced_aggregation.csv")
print(df) 
```

Calculate Total revenue per category

```python
revenue_per_category=df.groupby('Category')['Revenue'].sum()
print(revenue_per_category) 
```

 Calculate the Average units sold per product

```python

avg_units_per_product = df.groupby('Product')['Units_Sold'].mean()


print(avg_units_per_product)
```

Calculate Maximum customer rating per category

```python
max_rating_per_category=df.groupby('Category')['Customer_Rating'].max()
print(max_rating_per_category)  
```

Use groupby() and agg() to perform all calculations in one command.

```python

category_summary = df.groupby('Category').agg(
    Total_Revenue=('Revenue', 'sum'),
    Avg_Units_Sold=('Units_Sold', 'mean'),
    Max_Rating=('Customer_Rating', 'max')
)

print(category_summary)
```

Create a pivot table showing Total revenue per category

```python
pivot_table = df.pivot_table(
    index='Category',
    values='Revenue',
    aggfunc='sum'
)
print(pivot_table)
```

Create a pivot table showing Average customer rating per product

```python
pivot_table=df.pivot_table(
    index='Category',
    values='Customer_Rating',
    aggfunc='mean'

)
print(pivot_table)
```

Use pd.crosstab() to count how many times each product appears in each category.

```python
product_category_count= pd.crosstab(df['Product'], df['Category'])
print(product_category_count)
```

Identify the category with the highest total revenue

```python
revenue_per_category=df.groupby('Category')['Revenue'].sum()
highest_category=revenue_per_category.idxmax()
highest_revenue=revenue_per_category.max()
print(f"The category with the highest total revenue is '{highest_category}' with revenue {highest_revenue})")    
```
